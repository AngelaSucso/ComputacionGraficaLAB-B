### 6. Creen un programa que permita dibujar diferentes tipos de figuras geomtricas con el uso de eventos de mouse para poder hacer dibujos y luego guarden la imagen.

```python
import cv2
import numpy as np

imagen = cv2.imread('espacio.jpg')
type(imagen)

imagen_dibujo = imagen.copy()
def draw_figures(event, x, y, flags, param):
    global imagen_dibujo
    if event == cv2.EVENT_LBUTTONDOWN:
        imagen_dibujo = cv2.circle(imagen,(x+10,y+10),50,(0,255,0),-1)
    if event == cv2.EVENT_RBUTTONDOWN:
        imagen_dibujo = cv2.rectangle(imagen,(x,y),(x+100,y+100),(255,255,0),-1)
        
cv2.namedWindow(winname= 'drawing')
cv2.setMouseCallback('drawing', draw_figures)

while True:
    cv2.imshow('drawing',imagen_dibujo)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()

# GUARDA IMAGEN
cv2.imwrite('espacio_figuras.jpg',imagen_dibujo)
```
