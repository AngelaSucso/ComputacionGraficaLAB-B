2. Muestren solo el canal azul de la figura

```python
y = Image.open('messi.jpg')
y_array = np.asarray(y)
y_azul = y_array.copy()
y_azul[:,:,0] = 0
y_azul[:,:,1] = 0
plt.imshow(y_azul)
```

4. Guarden una nueva imagen que contenga solo los canales rojo y verde

```python
import cv2
imagen = cv2.imread('messi.jpg',1)
#BGR 0 = BLUE
imagen[:,:,0] = 0
cv2.imwrite('MessiSinAzul.png',imagen)
cv2.imshow('MessiPA',imagen)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
